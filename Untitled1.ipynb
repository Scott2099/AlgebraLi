```python
import matplotlib.pyplot as plt
import numpy as np
x= np.linspace(0,5,100)
y1= (3*x-4)/2
y2=(12-5*x)/2
plt.plot(x,y1,'-r',label='y1=(3x-4)/2')
plt.plot(x,y2,'--k',label='y2=(12-5*x)/2')
plt.title('GRafica de y_1 y y_2')
plt.xlabel('x')
plt.ylabel('y')
plt.legend(loc='upper left')
plt.grid()
```

```python
import matplotlib.pyplot as plt
import numpy as np
x= np.linspace(0,5,3,100)
y1= x-7
y2=(2*x-14)/2
plt.plot(x,y1,':r',label='y1=(x-7)',linewidth=7)
plt.plot(x,y2,'--k',label='y2=(2*x-14)/2',linewidth=3)
plt.title('GRafica de y_1 y y_2')
plt.xlabel('x',color='#1C2833')
plt.ylabel('y',color='#1C2833')
plt.legend(loc='upper left')
plt.grid()
```
